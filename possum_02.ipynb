```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Lectura de la base de datos

```python
datos = pd.read_csv('./possum.csv')
```

Para verificar la lectura de datos

```python
datos.head()
```

## 2. Operaciones con los datos

### 2.1 Eliminación de información no necesaria

```python
datos = datos.drop('case', axis=1)
```

```python
datos.head() # verificamos que case ya no existe
```

Información de los tipos de datos y su estructura

```python
datos.info()
```

### 2.2 Estadística básica sobre los datos

```python
datos.describe()
```

## 3. Visualización

```python
# datos individuales
datos['sex']
```

```python
# Graficas
datos['sex'].value_counts().plot(kind='bar')
```

### 3.1. Diagrama de bigotes

```python
fig, axs = plt.subplots(ncols=5, nrows=3, figsize=(20, 10))

index = 0
axs = axs.flatten()
for k,v in datos.items():
    sns.boxplot(y=k, data=datos, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

### 3.2 Distribución de los datos 

```python
fig, axs = plt.subplots(ncols=5, nrows=3, figsize=(20, 10))

index = 0
axs = axs.flatten()

for k, v in datos.items():
    sns.histplot(v, ax=axs[index], kde=True)
    axs[index].set_title(k)
    index += 1

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
plt.show()
```

### 3.2 Correlación

```python
sns.heatmap(datos.drop(['sex','Pop'], axis=1).corr(), cmap='coolwarm', annot=True)
```

### 3.3 Representación de los datos

```python
sns.scatterplot(data=datos, x='hdlngth', y='totlngth', hue='sex', size='age')
```
