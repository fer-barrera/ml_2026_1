```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Lectura de la base de datos

```python
datos = pd.read_csv('./possum.csv')
```

Para verificar la lectura de datos

```python
datos.head()
```

## 2. Operaciones con los datos

### 2.1 Eliminación de información no necesaria

```python
datos = datos.drop('case', axis=1)
```

```python
datos.head() # verificamos que case ya no existe
```

Información de los tipos de datos y su estructura

```python
datos.info()
```

## 3. Visualización

```python
# datos individuales
datos['Pop']
```

```python
# Graficas
datos['Pop'].value_counts().plot(kind='bar')
```

### 3.1. Diagrama de bigotes

```python
fig, axs = plt.subplots(ncols=5, nrows=3, figsize=(20, 10))

index = 0
axs = axs.flatten()
for k,v in datos.items():
    sns.boxplot(y=k, data=datos, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

### 3.2 Distribución de los datos 

```python
fig, axs = plt.subplots(ncols=5, nrows=3, figsize=(20, 10))

index = 0
axs = axs.flatten()

for k, v in datos.items():
    sns.histplot(v, ax=axs[index], kde=True)
    axs[index].set_title(k)
    index += 1

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
plt.show()
```

### 3.2 Correlación

```python
sns.heatmap(datos.drop(['sex','Pop'], axis=1).corr(), cmap='coolwarm', annot=True)
```

### 3.3 Visualización de los datos

## Relación entre longitud de cabeza y longitud total

Este comando genera un **gráfico de dispersión** que permite ver, de manera general, cómo se relacionan la **longitud de la cabeza** (`hdlngth`) y la **longitud total** (`totlngth`) de los individuos en la base de datos `datos`; además, utiliza el color (`hue='sex'`) para distinguir los puntos según el sexo y el tamaño (`size='age'`) para representar la edad, lo que ayuda a observar en una sola visualización si existe alguna tendencia entre ambas medidas y si esa relación cambia según el sexo o la edad.

```python
sns.scatterplot(data=datos, x='hdlngth', y='totlngth', hue='sex', size='age')
```

#### Relación entre longitud de cabeza y ancho del cráneo

Este comando crea un **gráfico de dispersión** para mostrar cómo se relacionan las variables `hdlngth` y `skullw` en el conjunto de datos `datos`; además, usa el color (`hue='sex'`) para diferenciar visualmente los puntos según el sexo, y el tamaño (`size='age'`) para representar la edad, de modo que en una sola gráfica se pueden observar al mismo tiempo la relación entre dos medidas físicas y cómo esa relación puede variar según el sexo y la edad de los individuos.

```python
sns.scatterplot(data=datos, x='hdlngth', y='skullw', hue='sex', size='age')
```

#### Relación entre variables con `pairplot`

El `pairplot` es una forma muy práctica y visual de “abstraer” los datos por primera vez, porque permite ver en una sola figura cómo se comporta cada variable por separado y también cómo se relaciona con las demás; así, sin hacer todavía un análisis profundo, se pueden notar tendencias, relaciones entre variables, grupos de datos y posibles valores atípicos, lo que lo convierte en una herramienta muy útil para entender rápidamente la estructura general de una base de datos.

```python
sns.pairplot(datos.drop(['sex','site'], axis=1), height=2.5, aspect=0.9)
```
