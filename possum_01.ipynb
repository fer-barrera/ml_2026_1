# Regresión con zarigüeyas

**Preámbulo**

## 1. Conjunto de datos de regresión de zarigüeyas

Las `zarigüeyas` son miembros del orden marsupial Didelphimorphia, endémico de América. Es el mayor orden de marsupiales del hemisferio occidental y comprende 93 especies distribuidas en 18 géneros.

**Contexto**

¿Puedes utilizar tus habilidades en regresión para predecir la edad de una zarigüeya, la longitud de su cabeza y si es macho o hembra? Este clásico conjunto de datos de práctica para regresión proviene originalmente del [paquete DAAG de R](https://cran.r-project.org/web/packages/DAAG/index.html) (datos usados en ejemplos y ejercicios del libro de Maindonald, J.H. y Braun, W.J. (2003, 2007, 2010), *Data Analysis and Graphics Using R*). Este conjunto de datos también se usa en el capítulo 8 del libro [OpenIntro Statistics](https://www.openintro.org/book/os/), *Introducción a la regresión lineal*.

**Contenido**

Del paquete DAAG de R: "*La datos de zarigüeyas consta de nueve mediciones morfométricas para cada una de 104 zarigüeyas de cola tupida de montaña, capturadas en siete sitios desde el sur de Victoria hasta el centro de Queensland*".

**Agradecimientos**

- Los datos se publicaron originalmente en el [paquete DAAG de R](https://cran.r-project.org/web/packages/DAAG/index.html) y se utilizaron en el libro de Maindonald, J.H. y Braun, W.J. (2003, 2007, 2010), *Data Analysis and Graphics Using R*.

- Un subconjunto de los datos también se reunió para el capítulo 8 del libro [OpenIntro Statistics](https://www.openintro.org/book/os/), *Introducción a la regresión lineal*.

***Fuente original del conjunto de datos:***

*Lindenmayer, D. B., Viggers, K. L., Cunningham, R. B., y Donnelly, C. F. 1995. Variación morfológica entre poblaciones de la zarigüeya de cola tupida de montaña, Trichosurus caninus Ogilby (Phalangeridae: Marsupiala). Australian Journal of Zoology 43: 449-458.*
