## 3. Carga de datos

Este comando carga en memoria una tabla de datos numéricos desde un archivo, asumiendo que los valores están separados por comas. Además, ignora la primera línea del archivo, que normalmente contiene los nombres de las columnas.

## 4. Matriz de correlación

Calcular la matriz de correlación entre las columnas

```python
# rowvar=False indica que cada columna es una variable
matriz_correlacion = np.corrcoef(datos, rowvar=False)
```

Imprime la matriz de correlación

```python
# Imprimir encabezado
print(" " * 8, end="")
for nombre in nombres_columnas:
    print(f"{nombre:>8}", end="")
print()

for i in range(len(nombres_columnas)):
    print(f"{nombres_columnas[i]:>8}", end="")
    for j in range(len(nombres_columnas)):
        print(f"{matriz_correlacion[i, j]:8.2f}", end="")
    print()
```

La **correlación** es una medida estadística que indica qué tan relacionadas están dos variables. La más común es la correlación de Pearson, la cual se calcula comparando cómo varían ambas variables respecto a sus medias y produce un valor entre -1 y 1. Un valor cercano a 1 indica una relación lineal positiva fuerte, es decir, cuando una variable aumenta la otra también tiende a aumentar; un valor cercano a -1 indica una relación lineal negativa fuerte, donde una sube y la otra baja; y un valor cercano a 0 indica que no existe una relación lineal clara entre ellas. Es importante tener en cuenta que correlación no implica causalidad.

La correlación de Pearson se calcula así:

$$
r_{xy} = \frac{\sum_{i=1}^{n}(x_i-\bar{x})(y_i-\bar{y})}
{\sqrt{\sum_{i=1}^{n}(x_i-\bar{x})^2}\sqrt{\sum_{i=1}^{n}(y_i-\bar{y})^2}}
$$

Donde $x_i$ y $y_i$ son los valores de las dos variables, $\bar{x}$ y $\bar{y}$ son sus medias, y $r_{xy}$ es el coeficiente de correlación.
